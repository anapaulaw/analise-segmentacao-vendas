```python
import sqlite3
import pandas as pd
```

```python
dados_brutos = pd.read_excel("vendas.xlsx")
```

```python
variaveis_desejadas = [
    'DATA E HORA',
    'NaturezaOperacao',
    'DEPARTAMENTO QUE VENDEU (OFICINA OU BALC√ÉO)',
    'CONDI√á√ÉO DE PAGAMENTO',
    'NFItem_PercDesc',
    'NFItem_QtdeEstoque',
    'NFItem_VlMargemCont',
    'NOME DA PE√áA',
    'VENDEDOR',
    'NFItem_VlDesc',
    'QUANTIDADE',
    'VALOR UNIT√ÅRIO',
    'NF_ProdCusto',
    'ProdPrecoValor',
    'ProdEstoqueClasABC',
    'ValorICMS',
    'ValorICMSST',
    'ValorPisCofins',
    'ValorLucroBruto',
    'VALOR DA VENDA',
    'NFItem_VlBruto',
    'NFItem_PercMargemCont',
    'NFItem_PercMargemGer'
]

df_filtrado = dados_brutos[variaveis_desejadas].copy()
```

### CRIAR BANCO SQLITE COM DADOS FILTRADOS

---



```python
conn = sqlite3.connect("vendas_filtradas.db")
df_filtrado.to_sql("vendas", conn, if_exists="replace", index=False)
print("‚úÖ Banco SQLite criado: vendas_filtradas.db")
```

```python
#SALVAR  os dados tamb√©m em Excel
df_filtrado.to_excel("vendas_filtradas.xlsx", index=False)
print("‚úÖ Arquivo Excel criado: vendas_filtradas.xlsx")
```

```python
consulta1 = pd.read_sql("""
    SELECT VENDEDOR, SUM("VALOR DA VENDA") AS Total_Vendas
    FROM vendas
    GROUP BY VENDEDOR
""", conn)
```

```python
consulta2 = pd.read_sql("""
    SELECT "DEPARTAMENTO QUE VENDEU (OFICINA OU BALC√ÉO)" AS Departamento,
           COUNT(*) AS Qtde_Vendas,
           SUM("VALOR DA VENDA") AS Total_Departamento
    FROM vendas
    GROUP BY Departamento
""", conn)



print("\nüîé Vendas por departamento:")
display(consulta2)
```
